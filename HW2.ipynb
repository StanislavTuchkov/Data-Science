```python
x = np.linspace(-3, 3, 51)
x
```

```python
y1 = x**2
y2 = 2*x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()

fig.set_size_inches(8, 6)
fig.subplots_adjust(wspace=0.3, hspace=0.3)

ax1.plot(x, y1, 'g')
ax1.set_xlim([-5, 5])
ax1.set_title('Schedule 1')
ax2.plot(x, y2, 'r--')
ax2.set_title('Schedule 2')
ax3.plot(x, y3, 'y^')
ax3.set_title('Schedule 3')
ax4.plot(x, y4, 'lightgrey')
ax4.set_title('Schedule 4')
```
